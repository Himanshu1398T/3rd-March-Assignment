### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### Ans:

1. Scatter Plot: A scatter plot is a two-dimensional plot that displays data points as dots on a coordinate system. It is useful for visualizing the relationship between two variables. It is often used to examine the correlation between two numerical variables.

2. Line Plot: A line plot displays data as a series of points connected by a straight line. It is useful for visualizing trends over time or for displaying data that has a natural order, such as categorical data. It is often used to examine the change of a variable over time.

3. Bar Plot: A bar plot is a chart that displays data using rectangular bars. It is useful for comparing different categories of data or for showing the distribution of data. It is often used to compare the frequency of occurrence of different categories.

4. Heat Map: A heat map is a graphical representation of data where the values are represented by colors. It is useful for visualizing the distribution of data over two or more dimensions. It is often used to identify patterns or relationships between variables.

5. Box Plot: A box plot is a graphical representation of the distribution of data based on the five-number summary: minimum, first quartile, median, third quartile, and maximum. It is useful for identifying outliers and for visualizing the spread of the data. It is often used to compare the distribution of data between different groups or categories.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

#### Note: timepoint, signal, event, and region are columns in the fmri dataset.

#### Ans:

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

fmri
```

```python
# Create a line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri).set_title('FMRI Dataset - Signal vs Timepoint')
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

#### Note: pclass, age, and fare are columns in the titanic dataset.

#### Ans:

```python
# Load the titanic dataset
titanic = sns.load_dataset("titanic")

titanic
```

```python
# Create a box plot for age by passenger class
sns.boxplot(x="pclass", y="age", data=titanic).set_title("Titanic Dataset - Age by Passenger Class")
```

```python
# Create a box plot for fare by passenger class
sns.boxplot(x="pclass", y="fare", data=titanic).set_title("Titanic Dataset - Fare by Passenger Class")
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
#### Ans:

```python
# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

diamonds
```

```python
# Create a histogram with the "price" column and the "cut" column as the hue parameter
sns.histplot(data=diamonds, x="price", hue="cut", kde=True, multiple="stack").set_title("Diamonds Dataset - Price Distribution by Cut")
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
#### Ans:

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.pairplot(data=iris, hue="species")
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
#### Ans:

```python
flights = sns.load_dataset("flights")
flights
```

```python
flights_matrix = flights.pivot("month", "year", "passengers")

flights_matrix
```

```python
sns.heatmap(flights_matrix)
```
